```
import plotly.express as px
```

```
df=pd.DataFrame(lloyds)
```

```
fig = px.treemap(df,
                 path=[px.Constant("all"), 'addr_state', 'loan_status', 'term'],
                 values='loan_amnt')

# Update the layout
fig.update_layout(margin=dict(t=50, l=25, r=25, b=25))

# Show the figure
fig.show()
```

```
import matplotlib.pyplot as plt

emp_title_counts = lloyds['emp_title'].value_counts()

# Select the top 10 employee titles based on count
top_10_emp_titles = emp_title_counts.head(10)

# Plotting the bar graph
plt.figure(figsize=(10, 6))
top_10_emp_titles.plot(kind='bar')
plt.title('Top 10 Employee Titles')
plt.xlabel('Employee Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```
emp_title_loan_status_counts = lloyds.groupby(['emp_title', 'loan_status']).size().unstack().fillna(0)

# Select the top 10 employee titles based on count
top_10_emp_titles = emp_title_loan_status_counts.sum(axis=1).sort_values(ascending=False).head(10)
top_10_emp_loan_status = emp_title_loan_status_counts.loc[top_10_emp_titles.index]

# Plotting the grouped bar graph
plt.figure(figsize=(12, 6))
top_10_emp_loan_status.plot(kind='bar', stacked=True)
plt.title('Top 10 Employee Titles and Loan Status')
plt.xlabel('Employee Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Loan Status')
plt.tight_layout()

# Display the plot
plt.show()
```
